```python
import pandas as pd
import dataUtilities

data = dataUtilities.parseData()
```

## Price by region

```python
df = pd.DataFrame(data, columns=['region', 'AveragePrice'])
df = df.groupby(['region']).mean().reset_index()
df = df.sort_values(by='AveragePrice', ascending=False)

values = df['AveragePrice'].values.tolist()
colors = [(0, x, 0) for x in dataUtilities.normalizeData(values)]

plot = df.plot(kind='bar', x='region', y='AveragePrice', figsize=(12,12), width=0.5, color=colors)
plot.set_title('Avocado Price by Region')
plot.set_xlabel('Region')
plot.set_ylabel('Average Price (USD)')
```

## Conventional vs Organic Price per Region

```python
df = pd.DataFrame(data, columns=['type', 'region', 'AveragePrice'])
df = df.groupby(['region', 'type']).mean().reset_index()

plot = df.pivot(index='region', columns='type', values='AveragePrice').plot(kind='bar', figsize=(12, 12), color=['DarkGreen', 'LightGreen'])
plot.set_title('Avocado Price - Conventional vs Organic by Region')
plot.set_ylabel('Average Price (USD)')
plot.set_xlabel('Region')
```

## Number of Avocados Sold vs Average Price

```python
df = pd.DataFrame(data, columns=['Total Volume', 'AveragePrice'])

plot = df.plot.hexbin(x='Total Volume', y='AveragePrice', figsize=(12,12), xscale='log')
plot.set_title('Avocado Price vs Volume')
plot.set_xlabel('Total # Of Avocados Sold')
plot.set_ylabel('Average Price (USD)')

```
